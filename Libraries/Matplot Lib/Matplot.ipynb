# Basic Plots


# Line plots:
there are useful for visualizing the relationship between
two variables over a continuous interval. To create a line plot,
use the plot() function in Matplotlib.

plt.xlabel('X-axis') sets the label for the x-axis to "X-axis".

plt.ylabel('Y-axis') sets the label for the y-axis to "Y-axis".

plt.title('Bar plot') sets the title of the plot to "graph plot".

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)
plt.xlabel('X-axis') # x-axis

plt.ylabel('Y-axis')

plt.title('Line Plot')

plt.show() # display the plot on the screen..
```

# Scatter Plots:
Scatter plots are ideal for displaying the distribution and
relationship between two numerical variables. Matplotlib
provides the scatter()

```python

x = [1,2,3,4,5]
y = [2,4,6,8, 10]

plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter plot')

plt.show()
```

# Bar Plots:
It is effective for comapring categorical data or showing the distribution of a single variable. Matplotlib offers the bar() and barh()
functions for craeting vertical and horizontal bar plots.

```python
categories = ['A', 'B', "C", "D"]
values = [10, 20, 15, 5]
plt.bar(categories, values)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar plot')

plt.show()
```

# Histograms:
It is useful for visualizing the distribution of continous variable. It provides the hist() function to create histograms..

```python
import numpy as np
data = np.random.randn(1000) # generate the data

plt.hist(data, bins=30) # bins: the range of data should be divided into 30 equal-width intervals, or bins.
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram plot')

plt.show()
```
